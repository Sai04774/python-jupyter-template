### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart in the first subplot
x = np.linspace(0, 10, 100)
y = x**2
axs[0, 0].plot(x, y, color='blue', label='y = x^2')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].legend()

# Bar chart in the second subplot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['green', 'orange', 'blue', 'red']
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot in the third subplot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='purple', label='Points')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].legend()

# Pie chart in the fourth subplot
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'orange', 'pink'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```
