### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

# Generate x values ranging from 0 to 10
x = np.arange(0, 11)  # Includes 0 to 10

# Compute y values as the square of x
y = x ** 2

# Create the plot
plt.plot(x, y, marker='o', label='y = x²')  # Added markers for better visibility

# Customize the plot
plt.title("Line Plot of y = x²", fontsize=14)
plt.xlabel("X values", fontsize=12)
plt.ylabel("Y values (square of X)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)  # Add a grid with a dashed style
plt.legend()  # Add a legend

# Display the plot
plt.show()




```
